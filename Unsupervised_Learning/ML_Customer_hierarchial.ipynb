```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
%matplotlib inline
```

```python
import warnings
warnings.filterwarnings("ignore", category=FutureWarning)
```

```python
os.chdir("D:\CDAC\Machine Learning")
df = pd.read_csv('Customer.csv')
```

```python
x=df.iloc[:,3:5].values
```

```python
import scipy.cluster.hierarchy as sch
# usind dendogram to find the optimal number of cluster
sch.dendrogram(sch.linkage(x,method='ward'))
plt.xlabel('Customers')
plt.ylabel('Euclideans distance')
plt.title('Dendrogram')
```

```python
# training model for hierarchical clustering

from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters = 5, affinity = 'euclidean',linkage='ward')
y_predhc = hc.fit_predict(x)
```

```python
y_predhc
```

```python
plt.scatter(x[y_predhc==0,0],x[y_predhc==0,1],s=10,c='red',label='Cluster 1')
plt.scatter(x[y_predhc==1,0],x[y_predhc==1,1],s=10,c='blue',label='Cluster 2')
plt.scatter(x[y_predhc==2,0],x[y_predhc==2,1],s=10,c='green',label='Cluster 3')
plt.scatter(x[y_predhc==3,0],x[y_predhc==3,1],s=10,c='cyan',label='Cluster 4')
plt.scatter(x[y_predhc==4,0],x[y_predhc==4,1],s=10,c='magenta',label='Cluster 5')
# plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=100,c='k',label='Centroid')
plt.legend()
```
