```python
x = new_data.iloc[:,0:13]

y = new_data.iloc[:,-1]
```

```python
from sklearn.impute import SimpleImputer
imputer_var = SimpleImputer(missing_values = np.nan, strategy = "mean")
imputer_var.fit(x)
x = imputer_var.transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.fit_transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm.fit(x_train,y_train)
```

```python
lm.coef_
```

```python
lm.intercept_
```

```python
lm_coeff = pd.Series(lm.coef_)
```

```python
lm_coeff.plot(kind ='bar')
```

```python
from sklearn.linear_model import Ridge
rid = Ridge()
rid.fit(x_train, y_train)
```

```python
rd_coeff = pd.Series(rid.coef_)
rd_coeff
```

```python
rd_coeff.plot(kind='bar')
```

```python
from sklearn.linear_model import Lasso
la = Lasso()
la.fit(x_train, y_train)
```

```python
la_coeff=pd.Series(la.coef_)
la_coeff
```

```python
la_coeff.plot(kind='bar')
```

```python
from sklearn.linear_model import ElasticNet
elnet = ElasticNet()
elnet.fit(x_train,y_train)
```

```python
elnet_coeff = pd.Series(elnet.coef_)
```

```python
elnet_coeff.plot(kind='bar')
```

```python
pred_lm = lm.predict(x_test)
pred_lm
```

```python
pred_rid = rid.predict(x_test)
pred_rid
```

```python
pred_la = la.predict(x_test)
pred_la
```

```python
pred_elnet = elnet.predict(x_test)
pred_elnet
```

```python
from sklearn import metrics
from sklearn.metrics import r2_score, mean_squared_error
metrics.r2_score(y_test,pred_lm)
```
